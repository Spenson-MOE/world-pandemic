## **Overview of the Diseases**<a id="diseases"></a>
[Up](#table)

**Polio** or poliomyelitis, is an infectious disease caused by the poliovirus. The poliovirus spreads through the saliva and feces of affected people. Most persons with polio have no symptoms, but others develop paralysis, which can be permanent or even fatal. Polio can cause mild flu-like symptoms to serious sickness damaging the spinal cord and brain. In its most extreme form, it might cause the inability to move specific limb.


**Guinea worm**, also known as dracunculiasis, is a parasitic infection caused by the nematode Dracunculus medinensis.  People become infected by drinking water contaminated with Guinea-worm larvae that reside inside copepods.  There is no specific medication for dracunculiasis. Treatment involves carefully wrapping the emerging worm around a small stick or gauze to encourage its exit. Efforts focus on filtering drinking water to remove copepods and educating people to avoid soaking affected limbs in sources of drinking water.


**Rabies** is a viral disease that causes encephalitis in humans and other mammals. It was historically referred to as hydrophobia (fear of water) due to the symptom of panic when presented with liquids to drink. Early symptoms can include fever and abnormal sensations at the site of exposure. These symptoms are followed by nausea, vomiting, violent movements, uncontrolled excitement, fear of water, an inability to move parts of the body, confusion, and loss of consciousness. Once symptoms appear, the result is virtually always death. Rabies is spread when an infected animal commonly dog bites or scratches a human or other animals.  Saliva from an infected animal can also transmit rabies if the saliva comes into contact with the eyes, mouth, or nose. Vaccination and rabies immunoglobulin are ways to prevent rabies. Washing bites and scratches for 15 minutes with soap and water, povidone-iodine, or detergent may reduce the number of viral particles and may be somewhat effective at preventing transmission.


**Malaria** is a mosquito-borne infectious disease caused by Plasmodium parasites. Malaria spreads through the bite of infected female Anopheles mosquitoes. Common symptoms include fever, chills, headache, muscle aches, and fatigue. Severe cases can lead to organ failure, coma, and death. Preventive measures include using insecticide-treated bed nets, taking antimalarial medications, and controlling mosquito populations. Prompt diagnosis and treatment with antimalarial drugs are essential.


**HIV** is a virus that attacks the bodys immune system. When someone is infected with HIV, they have HIV infection. HIV damages the immune system, making the body less able to fight infections and diseases.  AIDS is the most advanced stage of HIV infection. Not everyone with HIV develops AIDS. However, if left untreated, HIV infection typically progresses to AIDS over 10 to 15 years. AIDS severely weakens the immune system, making the person susceptible to opportunistic infections and certain cancers. Early diagnosis and treatment with antiretroviral drugs are crucial for managing HIV.


**Tuberculosis (TB)** is a serious infectious disease caused by the bacterium Mycobacterium tuberculosis. TB spreads through the air when infected individuals cough, sneeze, or spit. Tiny droplets containing TB germs can be inhaled by others, leading to infection. Some people experience flu-like symptoms like low fever, tiredness, and cough. When the immune system cant control the infection, germs cause disease in the lungs or other parts of the body. Symptoms include cough, chest pain, fever, weight loss, and fatigue. TB can also affect other organs (extrapulmonary TB). Symptoms vary based on the infected area. TB is preventable and curable. Antibiotics are used for treatment, but drug-resistant forms pose challenges.

```python
# Drop unwanted columns
used_data= df[['Entity','Code','Year',
                        "Total (estimated) polio cases",
        "Reported cases of guinea worm disease in humans",
        "Number of new cases of rabies, in both sexes aged all ages",
        "Number of new cases of malaria, in both sexes aged all ages",
        "Number of new cases of hiv/aids, in both sexes aged all ages",
        "Number of new cases of tuberculosis, in both sexes aged all ages",
               ]]
print_styled('_____________________________________________________')
print_styled("THE DATASET TO USE AFTER REMOVING UNWANTED COLUMNS")
print_styled('_____________________________________________________')
used_data.head(5)
```

```python
def print_bold(text):
    print(f"\033[1m{text}\033[0m")

def print_styled(text):
    # ANSI escape sequences for bold, red text with a yellow background
    print(f"\033[1;31;63m{text}\033[0m")
```

```python
# Select columns for disease case counts
# Update these column names based on your dataset structure
disease_columns = [
    "Total (estimated) polio cases",
    "Reported cases of guinea worm disease in humans",
    "Number of new cases of rabies, in both sexes aged all ages",
    "Number of new cases of malaria, in both sexes aged all ages",
    "Number of new cases of hiv/aids, in both sexes aged all ages",
    "Number of new cases of tuberculosis, in both sexes aged all ages",
]
```

```python
# Clean data: Replace NaN with 0
cleaned_columns=df[disease_columns].fillna(0)
clean_data = pd.concat([df[['Entity','Code','Year']], cleaned_columns],axis=1)
title= "\nDATA OF DISEASES IN THE WORLD\n"
print_styled('_____________________________________________')
print_styled(title)
print_styled('_____________________________________________')
clean_data.head(5)
```

```python
# Drop 'Entity' and 'Code' columns from clean_data to get Yearly trends so that i can use it to calculate globat trends
yearly_trends = clean_data.drop(columns=['Entity', 'Code'])

# Display the resulting yearly_trends DataFrame

```

## **Dashboard Visualization of Data analysed**<a id="dash"></a>

```python
# Set up the figure with 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
fig.tight_layout(pad=5.0)

# Bar Plot (Example: Sum of diseases per country)
year_data = clean_data[clean_data['Year'] == 2000]
top_10_countries = year_data.nlargest(10, "Number of new cases of malaria, in both sexes aged all ages")
sns.barplot(x='Entity', y="Number of new cases of malaria, in both sexes aged all ages", data=top_10_countries, ax=axes[0, 0])  # Example: 2nd column
axes[0, 0].set_title('Top 10 Countries with Highest Malaria Cases in 2000')
axes[0, 0].tick_params(axis='x', rotation=90)



# Plot each disease trend on the specified subplot
global_trends = yearly_trends.groupby('Year').sum()


# Plot each disease trend in the specified subplot (axes[0, 11])
axes[0, 1].plot(global_trends.index, global_trends[disease_columns], label=disease_columns)

# Set title, labels, and legend for the line plot
axes[0, 1].set_title('Disease Cases Trend', fontsize=14)
axes[0, 1].set_xlabel('Year', fontsize=12)
axes[0, 1].set_ylabel('Number of Cases', fontsize=12)
axes[0, 1].legend(loc='upper left', fontsize=10)



# Heatmap (Correlation of numerical columns)
sns.heatmap(cleaned_columns.corr(numeric_only=True), annot=True, cmap='coolwarm', ax=axes[1, 0])
axes[1, 0].set_title('Heatmap: Correlation')


# Regression plot between HIV and Year
sns.regplot(x='Year', y="Number of new cases of hiv/aids, in both sexes aged all ages", data=clean_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}, ax=axes[1,1])
axes[1,1].set_title('Regression Plot: Year vs HIV Cases')
axes[1,1].set_xlabel('Year')
axes[1,1].set_ylabel('HIV Cases')

#Display the dashboard
title="\nTHE DASHBOARD VISUALIZATION OF WORLD PANDEMIC RECORDS CASES\n"
print_styled('______________________________________________________________________')
print_styled(title)
print_styled('______________________________________________________________________\n')
print_styled(total_diseases)

avg_columns = df[disease_columns].mean().sort_values(ascending=False)
title="\nTHE AVERAGE RECORDS OF DISEASES CASES"
print_styled('\n_____________________________________________')
print_styled(title)
print_styled('_____________________________________________\n')
print_styled(avg_columns)
print('\n')
plt.show()
```

## **Insights and Recommendations**<a id="insights"></a>
[Up](#table)

```python
plt.figure(figsize=(12, 6))
plt.plot(global_trends.index, global_trends["Number of new cases of malaria, in both sexes aged all ages"], label="Malaria")
plt.annotate(
    "Malaria vaccine introduced",
    xy=(2019, global_trends["Number of new cases of malaria, in both sexes aged all ages"].loc[2019]),
    xytext=(2015, 1e6),
    arrowprops=dict(facecolor='red', shrink=0.05),
    fontsize=10,
    color="red"
)
plt.title("Highlighting Progress in Malaria Cases", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Cases", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
total_cases = global_trends.sum(axis=0)  # Sum of each disease column
percentage_change = (total_cases / total_cases.sum()) * 100
print_styled('____________________________________________')
print_styled("PERCENTAGE CHANGE OF EACH DISEASES:")
print_styled('____________________________________________\n')
print(percentage_change)
```

```python
# Plot histogram for progress (reductions in cases) and challenges (increases in cases)
mean_percentage = percentage_change.mean()
progress = percentage_change[percentage_change < mean_percentage]
challenges = percentage_change[percentage_change > mean_percentage]



# Combine progress and challenges into a single DataFrame for comparison
progress_challenges = pd.DataFrame({
    'Progress (Reduction)': abs(progress),
    'Challenges (Increase)': challenges
})

# Set up the figure
plt.figure(figsize=(20, 6))

# Histogram for progress (reductions)
plt.subplot(1, 2, 1)
progress.plot(kind='bar', color='green')
plt.title("Significant Progress in Diseases", fontsize=14, color='green', fontweight = 'bold')
plt.xlabel("Diseases", fontsize=12)
plt.ylabel("Average % Change (Reduction)", fontsize=12)
plt.xticks(rotation=90, fontsize=9)
plt.tight_layout()

# Histogram for challenges (increases)
plt.subplot(1, 2, 2)
challenges.plot(kind='bar', color='red')
plt.title("Persistent Challenges in Diseases", fontsize=14, fontweight='bold', color='red')
plt.xlabel("Diseases", fontsize=12)
plt.ylabel("Average % Change (Increase)", fontsize=12)
plt.xticks(rotation=90, fontsize=9)
plt.tight_layout()

# Display the histograms
plt.show()
```

1. Focus on Diseases with Significant Progress:
Diseases showing a reduction in cases, such as Malaria, can benefit from continued investment in successful prevention measures like vaccines, improved diagnostics, and better treatment plans. Efforts should be scaled up in regions where progress has been made to ensure further reductions.
Recommendation: Strengthen existing vaccination programs and expand awareness campaigns. Allocate more resources to regions showing positive trends to maximize progress.

2. Address Persistent Challenges:
Diseases like HIV/AIDS, Tuberculosis, and Malaria in certain regions show persistent increases in cases. This signals challenges in prevention, diagnosis, and treatment.
Recommendation: Increase investments in public health infrastructure, particularly in areas with high prevalence. Focus on strengthening health education, access to medications, and regular screening programs. Consider targeted interventions in areas with high disease burdens to curb increases in cases.

3. Monitor Trends Over Time:
Disease trends fluctuate over time, and external factors such as policy changes, availability of vaccines, and global events (e.g., pandemics like COVID-19) can significantly impact these trends.
Recommendation: Set up real-time monitoring systems that can track disease trends at a granular level (by country, region, and demographics). Use data-driven decision-making to adjust public health strategies promptly based on these trends.

4. Highlight Success Stories and Learn from Them:
The malaria vaccine's introduction and its potential to reduce malaria cases are a success story. This highlights the importance of vaccination and research in combating infectious diseases.
Recommendation: Leverage success stories to drive global health initiatives. Share best practices from regions where the vaccine or other successful interventions have been implemented to encourage adoption in other areas.

5. Invest in Education and Awareness Campaigns:
Many diseases, especially those showing persistent increases, are affected by societal behaviors, lack of awareness, and limited access to healthcare.
Recommendation: Launch targeted education campaigns that emphasize prevention, early diagnosis, and treatment. For diseases like HIV/AIDS, promoting safe practices and early testing is crucial.

6. Improvement in Data Accessibility and Analysis:
Data from global trends reveal important patterns and insights but may be underutilized in some regions.
Recommendation: Enhance accessibility to disease-related data and provide training for public health officials to analyze and interpret trends. This will help in making informed decisions at both the local and global levels.

7. Targeted Regional Interventions:
Disease trends can vary significantly by region. For example, some countries may show high rates of certain diseases, while others show little or no cases.
Recommendation: Tailor interventions based on regional disease burden. For instance, in regions with high rates of Malaria, focus on mosquito control and better access to antimalarial drugs, while in areas with rising HIV/AIDS rates, increase support for testing and treatment.

8. Collaboration with International Organizations:
Tackling global health issues requires collaboration between countries, health organizations, and research institutions.
Recommendation: Collaborate with global health organizations like the WHO, UNICEF, and non-governmental organizations to share resources, knowledge, and funding. Work together to develop cross-border strategies for managing infectious diseases.

9. Evaluate and Improve Vaccine Distribution:
The availability and distribution of vaccines can play a major role in reducing disease prevalence.
Recommendation: Focus on improving vaccine accessibility, particularly in low-income and high-risk areas. Address logistical challenges such as storage, transportation, and vaccine hesitancy to ensure widespread adoption.

10. Future Research into Emerging Diseases:
Emerging diseases may present new challenges in global health. Monitoring for new trends and understanding the impact of diseases in different populations is vital.
Recommendation: Increase funding for research on emerging diseases and invest in rapid-response mechanisms for new outbreaks. Set up global surveillance systems to track potential new pandemics or outbreaks as soon as they arise.
